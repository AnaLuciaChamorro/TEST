```python
from IPython.display import IFrame  
peliculasAnimadasExitosas="https://es.wikipedia.org/wiki/Anexo:Pel%C3%ADculas_animadas_con_las_mayores_recaudaciones" 
IFrame(peliculasAnimadasExitosas, width=900, height=900)
```

```python
import pandas as pd

tablaPeliculas=pd.read_html(peliculasAnimadasExitosas,
                        header=0, 
                        flavor='bs4')
```

```python
#observamos cuántas tablas hay en la página web
len(tablaPeliculas)
```

```python
#en este caso deseamos trabajar con la última tabla (tabla 8), es decir la 7
tablaPeliculas[7]
```

```python
# le asignamos un nombre a la tabla con la que trabajaremos
tablaPeliculaAnimadaMasExitosa=tablaPeliculas[7]
tablaPeliculaAnimadaMasExitosa
```

```python
tablaPeliculaAnimadaMasExitosa.info()
```

```python
#Pedimos el nombre de las columnas, para que posteriormente se nos haga más fácil cambiar el nombre de ciertas columnas
tablaPeliculaAnimadaMasExitosa.columns.to_list()
```

```python
#Cambiamos el nombre de dos columnas para que no existan símbolos en ellas
tablaPeliculaAnimadaMasExitosa.rename(columns={'N.º':'Num. de película','N.º de películas':'Cantidad de películas'},inplace=True)
tablaPeliculaAnimadaMasExitosa.columns
```

```python
#Para este trabajo nos resulta irrelevante la columna 'Num. de película' por eso la eliminaremos
byeColumns=['Num. de película']
tablaPeliculaAnimadaMasExitosa.drop(columns=byeColumns,inplace=True)
tablaPeliculaAnimadaMasExitosa
```

```python
#Creamos una copia de la tabla que ya tenemos
tablaNew=tablaPeliculaAnimadaMasExitosa.copy()
tablaNew
```

```python
tablaNew.columns.to_list()
```

```python
#Para no tener problemas con los nombres de las columnas renombramos algunos de ellas
tablaNew.rename(columns={'Recaudación a nivel mundial':'RecaudaciónMundial','Cantidad de películas':'PelículasProducidas','Promedio por película':'RecaudaciónPromedioPorPelícula','Película más exitosa de la serie':'PelículaMásExitosa'},inplace=True)
tablaNew.columns
```

```python
#Eliminamos los espacios en blanco de la columna 'RecaudaciónMundial'
tablaNew.RecaudaciónMundial=tablaNew.RecaudaciónMundial.str.replace(' ','')
```

```python
#Para ver si todos los espacios han sido eliminados
tablaNew.RecaudaciónMundial.to_list()
```

```python
#Eliminamos los espacios en blanco de la columna 'RecaudaciónPromedioPorPelícula'
tablaNew.RecaudaciónPromedioPorPelícula=tablaNew.RecaudaciónPromedioPorPelícula.str.replace(' ','')
```

```python
tablaNew.RecaudaciónPromedioPorPelícula.to_list()
```

```python
tablaPeliculas=tablaNew.copy()
tablaPeliculas
```

```python
#Eliminamos la columna 'RecaudaciónPromedioPorPelícula'
byeColumns=['RecaudaciónPromedioPorPelícula']
tablaPeliculas.drop(columns=byeColumns,inplace=True)
tablaPeliculas
```

```python
#Creamos una copia para guardar lo que ya tenemos
tablaPeliculasCopy=tablaPeliculas.copy()
tablaPeliculasCopy
```

```python
#Eliminamos lo datos que no son parte del nombre de la película
tablaPeliculasCopy['PelículaMásExitosa']=tablaPeliculasCopy['PelículaMásExitosa'].str.replace(r'\([^()]*\)','',regex=True)
tablaPeliculasCopy
```

```python
tablaPeliculasCopy.info()
```

```python
import os

tablaPeliculasCopy.to_csv(os.path.join("data","tablaPeliculasCopy.csv"),index=False)
```
