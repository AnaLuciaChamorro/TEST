```python
from IPython.display import IFrame  
tablaDisney="https://es.wikipedia.org/wiki/Anexo:Pel%C3%ADculas_de_Walt_Disney_Pictures" 
IFrame(tablaDisney, width=900, height=900)
```

```python
import pandas as pd

tablaPeliculas=pd.read_html(tablaDisney,
                        header=0, 
                        flavor='bs4')
```

```python
#para ver cuántas tablas hay en la página web
len(tablaPeliculas)
```

```python
#clasificación del estudio de grabación
tablaPeliculas[0]
```

```python
tablaPeliculas[10].info()
```

```python
tablaPeliculas[10].rename(columns={'Unnamed: 0':'Productor'},inplace=True)
tablaPeliculas[10].columns
```

```python
#elegir la tabla que queremos utilizar
tablaPeliculas[10]
```

```python
tablaElegida=tablaPeliculas[10].copy()
tablaElegida
```

```python
tablaElegida.Productor.str.strip()
```

```python
#cambiar el nombre de las filas de una columna
tablaElegida.Productor.replace({'D':'Walt Disney Pictures','W':'Walt Disney Animation Studios','P':'Pixar Animation Studios','T':'DisneyToon Studios','V':'Disney Television Animation','N':'Disneynature','D+':'Disney+','V+':'Disney Television Animation and Disney+','P+':'Pixar Animation Studios and Disney+'},inplace=True)
tablaElegida.head()
```

```python
tablaElegida
```

```python
tablaElegidaNew=tablaElegida.copy()
tablaElegidaNew
```

```python
#eliminar los espacios de las columnas
tablaElegidaNew.columns=tablaElegidaNew.columns.str.replace(' ','')
```
